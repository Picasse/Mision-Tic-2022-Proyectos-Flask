### Si-Sino-Finsi
Recuerda que los condicionales nos permiten decidir qué instrucciones ejecutar y cuales no. El condicional Si-Sino-Finsi tiene la siguiente estructura:

    Si (condición) Entonces
        instrucción(es)
    Sino
        instrucción(es)
    Fin Si

En Python, esto se escribe un poco diferente y la estructura general depende de las tabulaciónes. Por ejemplo:

```python
x = int(input("Por favor ingresa un número: "))
if x<0:
    print('X es Negativo')
else:
    print('X es Positivo')

print('Esto ya está por fuera del condicional') #Esta fuera porque ya no hay tabulación
```

¿Qué pasa si X es 0? ¿Imprimirá positivo o negativo? ¿Cómo podríamos arreglar este programa para que si X es cero, imprima "X es cero"?



```python
numero = int(input("Por favor ingresa un número: "))

if(numero>0):
  print("El número",numero,"es Positivo")

if(numero<0):
  print("El número",numero,"es Negativo")

if(numero==0):
  print("El número",numero,"es nulo")
```

Realizar el anterior programa con bloques anidados

```python
numero = int(input("Por favor ingresa un número: "))

if(numero>0):
  print("El número",numero,"es Positivo")
else:
  if (numero<0):
    print("El número",numero,"es Negativo")
  else:
    print("El número",numero,"es nulo")

print("Esto está fuera del bloque condional y en el bloque principal")
```

Realizar el programa anterior con la estructura if...elif..

```python
numero = int(input("Por favor ingresa un número: "))

if(numero>0):
  print("El número",numero,"es Positivo")
elif(numero<0):
  print("El número",numero,"es Negativo")
else:
  print("El número",numero,"es nulo")
```

Realizar un programa que dada una nota diga si el estudiante aprobó o no (aprueba con nota mayor o igual a 3.5) y si aprobó con nota entre 3.5 y 4 es un estudiante bueno, de 4 a 4.999 es sobresaliente y con 5 es excelente

```python
nota=float(input())
if(nota>=3.5):
    print("El estudiante aprobó")
    if (nota>=3.5 and nota<4):
      print("El estudiante es bueno")
    elif(nota>=4 and nota<5 ):
      print("El estudiante es sobresaliente")
    else:
      print("El estudiante es excelente")
else:
    print("El estudiante no aprobó")
```

Modificar el anterior programa para que no se digite una nota fuera del rango 0 a 5

```python
nota=float(input())
if (nota>=0 and nota <=5):
  if(nota>=3.5):
    print("El estudiante aprobó")
    if (nota>=3.5 and nota<4):
      print("El estudiante es bueno")
    elif(nota>=4 and nota<5 ):
      print("El estudiante es sobresaliente")
    else:
      print("El estudiante es excelente")
  else:
    print("El estudiante no aprobó")
else:
  print("Nota fuera de rango")
```
